# Function

```python
def create_barplot(data, x_col, y_col, title, xlabel, ylabel, top_n=20):
    """
    Function to create and display a bar plot.

    Parameters:
    - data: DataFrame with the relevant data.
    - x_col: Column name to be used for the x-axis (numeric values).
    - y_col: Column name to be used for the y-axis (categories).
    - title: Title for the plot.
    - xlabel: Label for the x-axis.
    - ylabel: Label for the y-axis.
    - top_n: Number of top countries to display (default is 20).
    """
    # Create a bar plot
    plt.figure(figsize=(10, 8))
    sns.barplot(data=data.sort_values(x_col, ascending=False).head(top_n)[::-1],
                x=x_col, y=y_col)
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.show()
```

# Analysis for Dataset 3

```python
cleaned_data = pd.read_csv('covid_19_clean_complete.csv', parse_dates=['Date'])
cleaned_data.head()
```

```python
# cases
cases = ['Confirmed', 'Deaths', 'Recovered', 'Active']

# Active Case = confirmed - deaths - recovered
cleaned_data['Active'] = cleaned_data['Confirmed'] - cleaned_data['Deaths'] - cleaned_data['Recovered']

# filling missing values
cleaned_data[['Province/State']] = cleaned_data[['Province/State']].fillna('')
cleaned_data[cases] = cleaned_data[cases].fillna(0)

cleaned_data.head()
```

```python
# Creating the DataFrames with total cases for every country
confirmed_cases = pd.DataFrame(cleaned_data.groupby('Country/Region')['Confirmed'].sum())
confirmed_cases['Country/Region'] = confirmed_cases.index
confirmed_cases.index = np.arange(1, len(confirmed_cases) + 1)

death_cases = pd.DataFrame(cleaned_data.groupby('Country/Region')['Deaths'].sum())
death_cases['Country/Region'] = death_cases.index
death_cases.index = np.arange(1, len(death_cases) + 1)

recovered_cases = pd.DataFrame(cleaned_data.groupby('Country/Region')['Recovered'].sum())
recovered_cases['Country/Region'] = recovered_cases.index
recovered_cases.index = np.arange(1, len(recovered_cases) + 1)

active_cases = pd.DataFrame(cleaned_data.groupby('Country/Region')['Active'].sum())
active_cases['Country/Region'] = active_cases.index
active_cases.index = np.arange(1, len(active_cases) + 1)

# Select relevant columns for each case
global_active_cases = active_cases[['Country/Region', 'Active']]
global_death_cases = death_cases[['Country/Region', 'Deaths']]
global_recovered_cases = recovered_cases[['Country/Region', 'Recovered']]
global_confirmed_cases = confirmed_cases[['Country/Region', 'Confirmed']]

# Create barplots for each category using the function

# For Confirmed Cases
create_barplot(global_confirmed_cases, 'Confirmed', 'Country/Region', 'Confirmed Cases Worldwide',
               'Confirmed Cases', 'Country/Region')

# For Death Cases
create_barplot(global_death_cases, 'Deaths', 'Country/Region', 'Death Cases Worldwide',
               'Deaths', 'Country/Region')

# For Recovered Cases
create_barplot(global_recovered_cases, 'Recovered', 'Country/Region', 'Recovered Cases Worldwide',
               'Recovered Cases', 'Country/Region')

# For Active Cases
create_barplot(global_active_cases, 'Active', 'Country/Region', 'Active Cases Worldwide',
               'Active Cases', 'Country/Region')
```

```python
cleaned_latest = cleaned_data[cleaned_data['Date'] == max(cleaned_data['Date'])]
flg = cleaned_latest.groupby('Country/Region')[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum().reset_index()

flg['mortalityRate'] = round((flg['Deaths'] / flg['Confirmed']) * 100, 2)

# Filter data where 'Confirmed' > 100 and sort by 'mortalityRate'
temp = flg[flg['Confirmed'] > 100]
temp = temp.sort_values('mortalityRate', ascending=False)

# Set the plot size
plt.figure(figsize=(10, 6))

# Create a horizontal bar plot with Seaborn, assigning 'mortalityRate' to 'hue'
sns.barplot(
    x='mortalityRate',
    y='Country/Region',
    data=temp.sort_values(by="mortalityRate", ascending=False)[:10][::-1],
    hue='Country/Region',  # Adding hue here
    dodge=False  # Removes legend and color variation for each country
)

# Set plot title and labels
plt.title('Deaths per 100 Confirmed Cases', fontsize=16, fontweight='bold')
plt.xlabel('Mortality Rate (%)')
plt.ylabel('Country/Region')

# Display the plot
plt.show()

```
